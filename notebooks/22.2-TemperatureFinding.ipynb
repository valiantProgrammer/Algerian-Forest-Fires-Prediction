## **__Algerian Forest Fires Dataset__** 
Data Set Information:

The dataset includes 244 instances that regroup a data of two regions of Algeria,namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012.
The dataset includes 11 attribues and 1 output attribue (class)
The 244 instances have been classified into fire(138 classes) and not fire (106 classes) classes.
Attribute Information:

1. Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012)
Weather data observations
2. Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
3. RH : Relative Humidity in %: 21 to 90
4. Ws :Wind speed in km/h: 6 to 29
5. Rain: total day in mm: 0 to 16.8
FWI Components
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
8. Drought Code (DC) index from the FWI system: 7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely Fire and not Fire

## **Test and Train Split**
I am taking Temperature as the output Feature and rest as input features

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.25, random_state=42)
```

```python
X_train.describe()
```

```python
X_train.corr()
```

```python
plt.Figure(figsize=(10,5))
sns.heatmap(X_train.corr(),annot=True,fmt='.2f')
plt.show()
```

```python
def correlation(dataset, threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold :
                colname= corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
drop_feature=correlation(X_train, 0.85)
drop_feature
```

```python
X_train.drop(drop_feature, axis=1, inplace=True)
X_test.drop(drop_feature, axis=1, inplace=True)
X_train.shape,X_test.shape
```

## **Standardization of the X_train and X_test**

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train_org = X_train
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train
```

```python
plt.subplots(figsize=(30,13))
plt.subplot(1,2,1)
sns.boxplot(data=X_train_org, legend="brief")
plt.title('X_train before Standardization')
plt.grid(False)
plt.subplot(1,2,2)
sns.boxplot(data=X_train)
plt.title('X_train after Standardization')
plt.grid(False)
plt.xticks(range(len(X_train_org.columns)), X_train_org.columns)
plt.show()
```

## **Linear Regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
linear_regression=LinearRegression()
linear_regression.fit(X_train,y_train)
y_pred = linear_regression.predict(X_test)
mae=mean_absolute_error(y_test, y_pred)
score=r2_score(y_test, y_pred)
print(mae)
print(score)
plt.scatter(y_test,y_pred)
plt.show()
```

## **Lasso Regression**

```python
from sklearn.linear_model import Lasso
lasso_regression=Lasso()
lasso_regression.fit(X_train,y_train)
y_pred = lasso_regression.predict(X_test)
mae=mean_absolute_error(y_test, y_pred)
score=r2_score(y_test, y_pred)
print(mae)
print(score)
plt.scatter(y_test,y_pred)
plt.show()
```

## **Ridge Regression**

```python
from sklearn.linear_model import Ridge
ridge_regression=Ridge()
ridge_regression.fit(X_train,y_train)
y_pred = ridge_regression.predict(X_test)
mae=mean_absolute_error(y_test, y_pred)
score=r2_score(y_test, y_pred)
print(mae)
print(score)
plt.scatter(y_test,y_pred)
plt.show()
```

## **Elastic Net Regression**

```python
from sklearn.linear_model import ElasticNet
elasticnet_regression=ElasticNet()
elasticnet_regression.fit(X_train,y_train)
y_pred = elasticnet_regression.predict(X_test)
mae=mean_absolute_error(y_test, y_pred)
score=r2_score(y_test, y_pred)
print(mae)
print(score)
plt.scatter(y_test,y_pred)
plt.show()
```

## **Cross Validation Lasso**

```python
from sklearn.linear_model import LassoCV
lassocv_regression=LassoCV(cv=5)
lassocv_regression.fit(X_train,y_train)
y_pred = lassocv_regression.predict(X_test)
mae=mean_absolute_error(y_test, y_pred)
score=r2_score(y_test, y_pred)
print(mae)
print(score)
plt.scatter(y_test,y_pred)
plt.show()
```

## **Cross Validation Ridge**

```python
from sklearn.linear_model import RidgeCV
lassocv_regression=RidgeCV(cv=5)
lassocv_regression.fit(X_train,y_train)
y_pred = lassocv_regression.predict(X_test)
mae=mean_absolute_error(y_test, y_pred)
score=r2_score(y_test, y_pred)
print(mae)
print(score)
plt.scatter(y_test,y_pred)
plt.show()
```
